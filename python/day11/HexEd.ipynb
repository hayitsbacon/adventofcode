```python
# https://adventofcode.com/2017/day/11

def move(direction, position):
    assert type(direction) is str
    assert type(position) is tuple
    if direction == 'n':
        # x y+1 z-1
        new_position = (position[0], position[1]+1, position[2]-1)
    if direction == 'ne': 
        # x+1 y z-1
        new_position = (position[0]+1, position[1], position[2]-1)
    if direction == 'se': 
        # x+1 y-1 z
        new_position = (position[0]+1, position[1]-1, position[2])
    if direction == 's': 
        # x y-1 z+1
        new_position = (position[0], position[1]-1, position[2]+1)
    if direction == 'sw': 
        # x-1 y z+1
        new_position = (position[0]-1, position[1], position[2]+1)
    if direction == 'nw': 
        # x-1 y+1 z
        new_position = (position[0]-1, position[1]+1, position[2])
    return new_position

with open('hexed.txt', 'r') as f:
    directions = f.read()
direction_list = directions.split(',')

# (x, y, z)
path = []
position = (0, 0, 0)
path.append(position)
for d in direction_list: 
    position = move(d, position)
    path.append(position)
```
